```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("datasets/logins.csv")
df.head()
```

```python
df.describe()
```

```python
# Invalid Lentgh Check
len_check = df["password"].str.len() >= 10
valid_pwd = df[len_check]
invalid_pwd = df[~len_check]
```

```python
valid_pwd.shape[0], invalid_pwd.shape[0]
```

```python
# Lower case check
lower_case = valid_pwd['password'].str.contains("[a-z]")
valid_pwd[lower_case].shape[0]
```

```python
# Upper case check
upper_case = valid_pwd['password'].str.contains("[A-Z]")
valid_pwd[upper_case].shape[0]
```

```python
# numeric check
numeric = valid_pwd['password'].str.contains("\d")
valid_pwd[numeric].shape[0]
```

```python
# special characteres check
special = valid_pwd['password'].str.contains("\W")
valid_pwd[numeric].shape[0]
```

```python
# comining all together
pd.concat([valid_pwd, lower_case, upper_case, numeric, special], axis=1)
```

```python
char_check = lower_case & upper_case & numeric & special

invalid_pwd = invalid_pwd.append(valid_pwd[~char_check], ignore_index=True)
valid_pwd = valid_pwd[char_check]
valid_pwd.shape[0], invalid_pwd.shape[0]
```

```python
# checking the pressence of the word "password" in the password
banned_pwd = valid_pwd["password"].str.contains("password", case=False)
invalid_pwd = invalid_pwd.append(valid_pwd[banned_pwd], ignore_index=True)
valid_pwd = valid_pwd[~banned_pwd]
valid_pwd.shape[0], invalid_pwd.shape[0]
```

```python
# checking the pressence of the first or last name in the password
valid_pwd["first_name"] = valid_pwd['username'].str.extract('(^[a-z]+)', expand= False)
valid_pwd["last_name"] = valid_pwd['username'].str.extract('([a-z]+$)')

valid_pwd
```

```python
for i, row in valid_pwd.iterrows():
    if row.first_name in row.password.lower() or row.last_name in row.password.lower() :
        print(row)
        valid_pwd = valid_pwd.drop(index=i)
        invalid_pwd = invalid_pwd.append(row, ignore_index=True)
```
